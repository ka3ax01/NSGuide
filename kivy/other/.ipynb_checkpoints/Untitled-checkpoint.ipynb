```python
Myu = np.array([])
Myu2 = np.array([])

a1, a2, count = float('0.0'), float('0.0'), 0
el1, el2 = 0, 0
check = 0
for space in Z:
    if check == 0:
        #a1 = a1 + float(space)
        #el1 = float(space)
        Myu = np.append(Myu, space)
        check = 1
    elif check == 1:
        #a2 = a2 + float(space)
        #el2 = float(space)
        Myu2 = np.append(Myu2, space)
        check = 2
    elif check == 2:
        check=0
    count = count +1
Z = np.concatenate(([Myu], [Myu2]), axis=0)

plot(Z[0,:],Z[1,:],'.')
axis((-50,40,-40,40))
show()
```
